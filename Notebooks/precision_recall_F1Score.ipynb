```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
```

```python
dataset=pd.read_csv('heart.csv')
```

```python
dataset.head()
```

```python
x=dataset.drop(columns='target',axis=1)
y=dataset.target
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

## Accuracy Score 

```python
from sklearn.metrics import accuracy_score
```

```python
#accuracy score on the testing data is 
train_pre=model.predict(x_train)
train_acc=accuracy_score(y_train,train_pre)
```

```python
print("The accuracy score on the training data is :",round(train_acc*100,2),"%")
```

```python
#the accuracy on testing data is
test_pre=model.predict(x_test)
test_acc=accuracy_score(y_test,test_pre)
```

```python
print("the accuracy score on the testing data is :",round(test_acc*100,2),'%')
```

## Precision 

```python
from sklearn.metrics import precision_score
```

```python
#precision for training data predictions 
precision_train=precision_score(y_train,train_pre)
print("training data precision is :",round(precision_train*100,2),"%")
```

```python
#precision for the testing data 
precision_test=precision_score(y_test,test_pre)
print("testing data precision is:",round(precision_test*100,2),"%")
```

## Recall Score

```python
from sklearn.metrics import recall_score
```

```python
#recall for training data predictions 
recall_train=recall_score(y_train,train_pre)
print("training data recall is :",round(recall_train*100,2),"%")
```

```python
#recall for the testing data 
recall_test=recall_score(y_test,test_pre)
print("testing data recall is:",round(recall_test*100,2),"%")
```

## F1 Score

```python
from sklearn.metrics import f1_score
```

```python
#f1 socre for training data predictions 
f1_train=f1_score(y_train,train_pre)
print("training data f1_score is :",round(f1_train*100,2),"%")
```

```python
#f1score for the testing data 
f1_test=f1_score(y_test,test_pre)
print("testing data f1_score is:",round(f1_test*100,2),"%")
```

## function for precision recall and f1score

```python
def precision_recall_f1score(true_label,predicted_label):
        precision_value=precision_score(true_label,predicted_label)
        recall_value=recall_score(true_label,predicted_label)
        f1_score_value=f1_score(true_label,predicted_label)
        print('precision:',round(precision_value*100,2))
        print('recall:',round(recall_value*100,2))
        print('f1_score:',round(f1_score_value*100,2))
    
```

```python
#classification matrix for training data is 
precision_recall_f1score(y_train,train_pre)
```
