```python
#importing the dependencies 
import numpy as np 
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
#importing different types of model
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#loading the dataset
dataset=pd.read_csv('heart.csv')
```

```python
dataset.isnull().sum()
```

```python
dataset['target'].value_counts()
```

```python
x=dataset.drop(columns='target',axis=1)
y=dataset.target
```

```python
print(x)
```

```python
print(y)
```

```python
X=np.asarray(x)
Y=np.asarray(y)
```

### Model Selection 

1.Compare the model with default hyperparameter value using cross validation 

```python
#list of models
models=[LogisticRegression(max_iter=1000),SVC(kernel='linear'), KNeighborsClassifier(),RandomForestClassifier(random_state=0)]
```

```python
def compare_model_cross_validation():
    for model in models:
        cv_score=cross_val_score(model,X,Y,cv=5)
        mean_accuracy=sum(cv_score)/len(cv_score)
        print(f"the cross validation score of {model} is ={cv_score}")
        print(f"the accuracy score of the {model} is:{mean_accuracy*100:.2f}%")
        print("-------------------------------------------------------------")
```

```python
compare_model_cross_validation()
```

**Inference**:The linear regression has the highset accuracy score & the kNeighborclassifier has the lowest accuracy score 

### Comparing the model with different hyperparameter using GridSearchCV

```python
#list of models
model_list=[LogisticRegression(max_iter=10000),SVC(), KNeighborsClassifier(),RandomForestClassifier(random_state=0)]
```

```python
#creating the dictonary containing all the values for above mention models 
model_hyperparameter={
    'log_reg_hyperparameters':{
    'C':[1,5,10,20]
    
    },
    'svc_hyperparameters':{
        'kernel':['linear','poly','rbf','sigmoid'],'C':[1,5,10,20]
    },
    'KNN_hyperparameter':{
        'n_neighbors':[3,5,10]
    },
    'random_forest_hyperparameters':{
        
        'n_estimators':[10,20,50,100]
    }
    
}
```

## Applying the GridSreachCV

```python
 def ModelSelection(list_of_models,hyperparameter_dict):
     result=[]
     i=0
     for model in list_of_models:
             key=model_keys[i]
             params=hyperparameter_dict[key]
             i+=1
             print(model)
             print(params)
             print('---------------------------------------------------')
        #appyling the girdsearchcv
             classifier=GridSearchCV(model,params,cv=5)
             #fitting the data
             classifier.fit(X,Y)
             result.append({
                 'model used':model,
                 'highest score':classifier.best_score_*100,
                 'best hyperparameter':classifier.best_params_
             })
     result_dataframe=pd.DataFrame(result,columns=['model used','highest score','best hyperparameter'])
     return result_dataframe
    
            
 
```

```python
ModelSelection(model_list,model_hyperparameter)
```

**RandomForestClassifier with n_estimators:100 has the highest accuracy among all the models**
