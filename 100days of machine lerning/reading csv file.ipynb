# Converters 

```python
#suppose i have to change the name in the dataset from royal challengers banglore to rcb what i will do is 
# i will define the function
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return "RCB"
    else:
        return name

```

```python
# lets apply this into a dataset
ipl=pd.read_csv('../dataset/IPL Matches 2008-2020.csv',converters={'team1':rename})
```

```python
ipl
```
