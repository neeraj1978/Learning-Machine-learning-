## Confusion matrix 

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#loading the dataset
dataset=pd.read_csv('heart.csv')
```

```python
dataset.head()
```

```python
x=dataset.drop(columns='target',axis=1)
y=dataset.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
#accurcy score 
x_train_pre=model.predict(x_train)
x_train_acc=accuracy_score(y_train,x_train_pre)
```

```python
print(f"the accuracy score on training data is:{x_train_acc}")
```

```python
#accuracy score on testing data
x_test_pre=model.predict(x_test)
x_test_acc=accuracy_score(y_test,x_test_pre)
```

```python
print(f"the accuracy score on testing data is:{x_test_acc}")
```

## confusion Matrix 

it is totally based on the testing data so we will be making it on the testing data

```python
cf_matrix=confusion_matrix(y_test,x_test_pre)
print(cf_matrix)
```

```python
tn,fp,fn,tp=cf_matrix.ravel()
print(tn,fp,fn,tp)
```

```python
import seaborn as sns
sns.heatmap(cf_matrix,annot=True)
```
