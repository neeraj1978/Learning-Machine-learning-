### Timer

```python
import time 
```

```python
time.sleep(3)
print("time's up!")
```

```python
count=int(input("enter the countdown timer"))
time.sleep(count)
print("Time's Up!") 
```

```python
count=int(input("enter the countdown timer"))
for x in range(0,count):
    print(x)
    time.sleep(1)
print("Time's UP!")
```

```python
count=int(input("enter the countdown timer"))
for x in reversed(range(0,count)):
    print(x)
    time.sleep(1)
print("Time's UP!")
```

```python
count=int(input("enter the countdown timer"))
for x in reversed(range(0,count)):
    seconds=x%60
    minutes=int((x/60)%60)
    hours=int(x/3600)
    print(f"{hours:02}:{minutes:02}:{seconds:02}")
    time.sleep(1)
print("time's up")
   
```
