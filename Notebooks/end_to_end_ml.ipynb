```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score 
```

```python
df=pd.read_csv('../dataset/placement.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.iloc[:,1:]
```

```python
df.head()
```

```python
#making the scatter plot 
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
print(x)
```

```python
print(y)
```

```python
#train test split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

```python
x_test
```

```python
#scaling the values 
scaler=StandardScaler()
```

```python
scaler.fit_transform(x_train)
```

```python
scaler.fit_transform(x_test)
```

```python
#loading the model
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#accuracy score on training data
train_pre=model.predict(x_train)
train_acc=accuracy_score(y_train,train_pre)
```

```python
print("The accuracy score on training data is",train_acc)
```

```python
#accuracy score on testing data
test_pre=model.predict(x_test)
test_acc=accuracy_score(y_test,test_pre)
```

```python
print("The accuracy score on testing data is",test_acc)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train.values, y_train.values, clf=model, legend=2)
```
